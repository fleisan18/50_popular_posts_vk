3. Выгрузим в переменную `posts` в формате датафрейма __все записи__ сообщества Skyeng: https://vk.com/skyeng, 

```python
import time
```

```python
offset = 0
max_count = 100
posts = pd.DataFrame()

while True:      # Повторять цикл пока его не остановят по условию ниже
    
    url = 'https://api.vk.com/method/wall.get'
    params = {
        'owner_id': '-' + str(51890028),
        'access_token': token,
        'v':'5.131',
        'filter': 'all',
        'count': max_count,
        'offset': offset
    }
    
    req = requests.get(url, params)
    res = req.json()
    
    if len(res['response']['items']) != 0:      # Проверяем вернулись ли нам хоть какие-нибудь записи
        posts = pd.concat([posts, pd.DataFrame(res['response']['items'])]) 
    else:
        break                                   # Если нам ничего не вернулось, то прервать цикл
    print('Уже отображено', offset, 'постов')
    offset = offset + max_count                 # Смещаем offset на число возвращенных записей
    time.sleep(0.5)                             # Опять контролируем частоту запросов, чтобы нас не ограничили

```

```python
print('Кол-во записей:', len(posts))
```

__4. Выгрузим 50 самых популярных постов. Метрикой популярности мы считаем сумму лайков, репостов и комментариев.__

1. Создадим столбцы с числовыми значениями для лайков, репостов и комментариев и заберем из словаря только значение по ключу `count` 
2. Создадим отдельный столбец, где будет сумма пользовательских активностей для каждого поста
3. Отсортируем дата-фрейм по столбцу с суммой пользовательских активностей
4. Выберем 50 самых популярных постов

```python
posts.columns
```

```python
# вытащим из словарей столбцов comments, likes, reposts только нужные значения

activity = posts[['comments', 'likes', 'reposts']]
activity['comments_count'] = activity['comments'].apply(lambda x: x.get('count'))
activity['likes_count'] = activity['likes'].apply(lambda x: x.get('count'))
activity['reposts_count'] = activity['reposts'].apply(lambda x: x.get('count'))
activity

```

```python
# заполним даными с прошлого шага новые столбцы
posts['comments_count'] = activity['comments_count']
posts['likes_count'] = activity['likes_count']
posts['reposts_count'] = activity['reposts_count']
```

```python
# заполним данными столбец с суммой по всем активностям

posts['activity'] = posts['comments_count'] +posts['likes_count'] + posts['reposts_count']
posts
```

```python
# сортировка
posts.sort_values(by='activity', ascending = False, inplace = True)
posts
```
